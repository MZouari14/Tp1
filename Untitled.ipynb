```python
df['Latitude totale'] = df['LatD'] + (df[' "LatM"'] / 60) + (df[' "LatS"'] / 3600)
df['Longitude totale'] = df[' "LonD"'] + (df[' "LonM"'] / 60) + (df[' "LonS"'] / 3600)
```

```python

df.columns = df.columns.str.replace('"', '', regex =False)
df[' NS'] = df[' NS'].str.replace('"', '').str.strip()
df.columns = df.columns.str.strip()

print(df.columns)
print(df.head())
```

```python
# Filtrer pour obtenir uniquement les villes de l'hémisphère Nord
if 'NS' in df.columns:
    df_nord = df[df['NS'] == 'N']
    
    # Afficher le DataFrame filtré
    print(df_nord.head())
else:
    print("La colonne 'NS' est manquante.")
```

```python
df_population = pd.read_csv('C:/Users/MelekZOUARI/Kuikops_TP1/Population_data.csv')

# Vérifier les colonnes disponibles
print(df.columns)
print(df_population.columns)
# Fusionner les DataFrames sur la colonne "City"
df_merged = pd.merge(df, df_population, on='City', how='left')

# Afficher les premières lignes du DataFrame fusionné
print(df_merged.head())
```

```python
df_merged['Population_2025'] = df_merged['Population_2020'] * (1 + df_merged['Growth_rate'] / 100) ** 5
    
print(df_merged[['City', 'Population_2020', 'Growth_rate', 'Population_2025']].head())

```

```python
# Filtrer les villes avec une Population_2025 > 1 million
villes_1_million = df_merged[df_merged['Population_2025'] > 1_000_000]

# Afficher les villes correspondantes
print(villes_1_million[['City', 'Population_2020', 'Growth_rate', 'Population_2025']])

villes_1_million

```

```python
import matplotlib.pyplot as plt

# 1. Trier le DataFrame par la colonne Population_2025 en ordre décroissant
top_10_villes = df_merged.sort_values(by='Population_2025', ascending=False).head(10)

# 2. Créer un graphique à barres
plt.figure(figsize=(10, 6))
plt.barh(top_10_villes['City'], top_10_villes['Population_2025'], color='cyan')
plt.xlabel('Population Projetée en 2025')
plt.ylabel('Ville')
plt.title('Top 10 des villes avec la population projetée la plus élevée en 2025')
plt.gca().invert_yaxis()  # Inverser l'axe des y pour avoir la plus grande valeur en haut

# 3. Afficher le graphique
plt.show()
```
